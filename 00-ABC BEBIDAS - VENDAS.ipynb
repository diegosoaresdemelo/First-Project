## 1. 0 - DADOS

### 1.1 IMPORTAÇÃO DOS DADOS E LIMPEZA DOS DADOS

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#DADOS MÊS DE JANEIRO ATÉ MARÇO
vendasJanMar = pd.read_excel("https://github.com/diegosoaresdemelo/First-Project/blob/master/00-VENDAS%20ABC%20-%20JAN-MAR.xlsx?raw=true")
vendasJanMar
```

```python
#DADOS MÊS DE ABRIL ATÉ MAIO
vendasAbrMai = pd.read_excel("https://github.com/diegosoaresdemelo/First-Project/blob/master/01-VENDAS%20ABC%20-%20ABR-MAI.xlsx?raw=true")
vendasAbrMai
```

```python
#JUNTAR AS INFORMAÇÕES DE VENDA
vendasGeral = vendasJanMar.append(vendasAbrMai)
vendasGeral
```

```python
#IMPORTAÇÃO DOS DADOS DAS LOJAS
lojas = pd.read_excel("https://github.com/diegosoaresdemelo/First-Project/blob/master/02-VENDAS-ABC-LOJAS.xlsx?raw=true")
lojas.head()
```

```python
#JUNTAS AS INFORMAÇÃO DAS VENDAS CONFORME LOJA
vendasGeral = vendasGeral.merge(lojas)
vendasGeral.head()
```

```python
#VERIFICAR LINHAS VAZIAS E COLUNAS SEM INFORMAÇÃO
vendasGeral.info()
```

```python
# EXCLUIR AS COLUNAS SEM RELEVÂNCIA PARA ANÁLISE
vendasGeral = vendasGeral.drop(columns=["Unnamed: 3"])
```

```python
#EXCLUIR AS LINHAS VAZIAS SEM RELEVÂNCIA PARA ANÁLISE 
vendasGeral = vendasGeral.dropna()
```

```python
#VERIFICAR NOVAMENTE AS INFORMAÇÕES APÓS LIMPEZA DOS DADOS
vendasGeral.info()
```

```python
#CRIAÇÃO DE COLUNA P/ A MULTIPLICAÇÃO DOS PRODUTOS VENDIDOS x PREÇO DE VENDA PARA DESCOBRIR O FATURAMENTO
vendasGeral["Total Vendido"] = vendasGeral["Quant.Vendida"]*vendasGeral["Preço de Venda"]
vendasGeral.head()
```

## 2.0 LEVANTAMENTO DAS INFORMAÇÕES

### 2.1 FATURAMENTO GERAL DOS PRODUTOS VENDIDOS
### 2.2 QUANTIDADE GERAL DOS PRODUTOS VENDIDOS

```python
#SOMA DA QUANTIDADE VENDIDA TOTAL NO BRASIL E DO FATURAMENTO TOTAL DO BRASIL
faturamentoGeral = (vendasGeral[["Quant.Vendida","Total Vendido"]].sum()).to_frame()
faturamentoGeral = faturamentoGeral.rename(columns={0:"QUANT. VENDIDA x FATURADO"})
faturamentoGeral
```

```python
#GRÁFICOS DE BARRA HORIZONTAL PARA REPRESENTAR FATURAMENTO X VENDIDO
faturamentoGeral.plot.barh()
plt.title("FATURAMENTO TOTAL x QUANTIDADE TOTAL - BRASIL")
sns.set()
```

### 2.3 CATEGORIA COM MAIOR QUANTIDADE DE PRODUTO VENDIDO - BRASIL
### 2.4 CATEGORIA COM MAIOR FATURAMENTO DE PRODUTO  - BRASIL

```python
#SOMA DA QUANTIDADE DE PRODUTO VENDIDO POR CATEGORIA
categoriaQuantidade = vendasGeral[["Categoria","Quant.Vendida"]].groupby("Categoria").sum().sort_values(by="Quant.Vendida",ascending=False)
```

```python
categoriaQuantidade
```

```python
#SOMA DA QUANTIDADE DE PRODUTO FATURADO POR CATEGORIA
faturamentoCatTotal = vendasGeral[["Categoria","Total Vendido"]].groupby("Categoria").sum().sort_values(by="Total Vendido",ascending=False)
```

```python
faturamentoCatTotal
```

```python
#GRÁFICO DE BARRA HORIZONTAL PARA REPRESENTAR QUANTIDADE VENDIDA POR CATEGORIA
categoriaQuantidade.plot.barh()
plt.title("CATEGORIA COM MAIOR QUANTIDADE DE PRODUTO VENDIDO - BRASIL")
plt.xlabel("Quantidade Vendida")
plt.ylabel("Categoria")
#GRÁFICO DE BARRA HORIZONTAL PARA REPRESENTAR QUANTIDADE FATURADA POR CATEGORIA
faturamentoCatTotal.plot.barh()
plt.title("CATEGORIA COM MAIOR FATURAMENTO DE PRODUTO VENDIDO - BRASIL")
plt.xlabel("Valor Faturado")
plt.ylabel("Categoria")
sns.set()
```

### 2.5 TOP 5 PRODUTOS VENDIDOS - BRASIL
### 2.6 TOP 5 PRODUTOS FATURADOS - BRASIL

```python
#SOMA DA QUANTIDADE VENDIDA POR PRODUTO - TOP 5 BRASIL
produtoVendido = vendasGeral[["Descrição do Produto","Quant.Vendida"]].groupby("Descrição do Produto").sum().sort_values(by="Quant.Vendida",ascending=False).head(5)
produtoVendido
```

```python
#SOMA DA QUANTIDADE FATURADA POR PRODUTO - TOP 5 BRASIL
produtoFaturado = vendasGeral[["Descrição do Produto","Total Vendido"]].groupby("Descrição do Produto").sum().sort_values(by="Total Vendido",ascending=False).head(5)
produtoFaturado
```

```python
#GRÁFICO DE BARRA HORIZONTAL PARA REPRESENTAR QUANTIDADE VENDIDA POR PRODUTO - TOP 5 BRASIL
produtoVendido.plot.barh()
plt.title("TOP 5 PRODUTOS VENDIDOS - BRASIL")
plt.xlabel("Quantidade Vendida")
plt.ylabel("Produto")
#GRÁFICO DE BARRA HORIZONTAL PARA REPRESENTAR FATURADA VENDIDA POR PRODUTO - TOP 5 BRASIL
produtoFaturado.plot.barh()
plt.title("TOP 5 PRODUTOS FATURADOS - BRASIL")
plt.xlabel("Valor Faturado")
plt.ylabel("Produto")
sns.set()
```

### 2.7 CATEGORIA COM MAIOR QUANTIDADE DE PRODUTO VENDIDO - ESTADO

```python
#LISTA DAS CATEGORIAS
categoria = vendasGeral["Categoria"].unique()
```

```python
#SOMA DA QUANTIDADE VENDIDA POR CATEGORIA EM CADA ESTADO
categoriaQuantEstado = pd.crosstab(vendasGeral["UF Loja"],vendasGeral["Categoria"],values=vendasGeral["Quant.Vendida"],aggfunc="sum")
categoriaQuantEstado = categoriaQuantEstado.fillna(0)
categoriaQuantEstado["Total Vendido"] = categoriaQuantEstado["Água"]+categoriaQuantEstado["Chá"]+categoriaQuantEstado["Refrigerante"]+categoriaQuantEstado["Energético"]
categoriaQuantEstado = categoriaQuantEstado.sort_values(by="Total Vendido",ascending=False)
categoriaQuantEstado
```

```python
#GRÁFICO DE BARRA HORIZONTAL DA QUANTIDADE VENDIDA POR CATEGORIA EM CADA ESTADO
for categ in categoria:
    categoriaQuantEstGraf = vendasGeral.loc[vendasGeral["Categoria"]==categ,["UF Loja","Quant.Vendida"]].groupby("UF Loja").sum().sort_values(by="Quant.Vendida",ascending=True)
    categoriaQuantEstGraf.plot.barh()
    plt.title(f"QUANTIDADE TOTAL VENDIDA POR CATEGORIA - {categ}")
    plt.xlabel("Quantidade Vendida")
    plt.ylabel("ESTADO")
    sns.set()
```

### 2.8 CATEGORIA COM MAIOR FATURAMENTO DE PRODUTO - ESTADO

```python
#SOMA DA QUANTIDADE FATURADA POR CATEGORIA EM CADA ESTADO
categoriaFaturaEstado = pd.crosstab(vendasGeral["UF Loja"],vendasGeral["Categoria"],values=vendasGeral["Total Vendido"],aggfunc="sum")
categoriaFaturaEstado = categoriaFaturaEstado.fillna(0)
categoriaFaturaEstado["Total Faturado"] = categoriaFaturaEstado["Água"]+categoriaFaturaEstado["Chá"]+categoriaFaturaEstado["Refrigerante"]+categoriaFaturaEstado["Energético"]
categoriaFaturaEstado = categoriaFaturaEstado.sort_values(by="Total Faturado",ascending=False)
categoriaFaturaEstado
```

```python
#GRÁFICO DE BARRA HORIZONTAL DA QUANTIDADE FATURADA POR CATEGORIA EM CADA ESTADO
for categ in categoria:
    categoriaFaturaEstGraf = vendasGeral.loc[vendasGeral["Categoria"]==categ,["UF Loja","Total Vendido"]].groupby("UF Loja").sum().sort_values(by="Total Vendido",ascending=True)
    categoriaFaturaEstGraf.plot.barh()
    plt.title(f"FATURAMENTO TOTAL POR CATEGORIA - {categ}")
    plt.xlabel("Quantidade Vendida")
    plt.ylabel("ESTADO")
    sns.set()
```

### 2.9 TOP 5 PRODUTOS VENDIDOS - ESTADO

```python
#LISTA DOS ESTADOS
estado = vendasGeral["UF Loja"].unique()
```

```python
#GRÁFICO DE BARRA HORIZONTAL DA SOMA DA QUANTIDADE VENDIDA POR PRODUTO - TOP 5 ESTADO
for estados in estado:
    topProdutosEstado = vendasGeral.loc[vendasGeral["UF Loja"]==estados,["UF Loja","Descrição do Produto","Quant.Vendida"]].groupby("Descrição do Produto").sum().sort_values(by="Quant.Vendida",ascending=False).head(5)
    topProdutosEstado.plot.barh()
    plt.title(f"TOP 5 PRODUTOS VENDIDOS - {estados}")
    plt.xlabel("Quantidade Vendida")
    plt.ylabel("PRODUTOS")
    sns.set()
```

### 2.10 TOP 5 FATURAMENTO - ESTADO

```python
#GRÁFICO DE BARRA HORIZONTAL DA SOMA DA QUANTIDADE FATURADA POR PRODUTO - TOP 5 ESTADO
for estados in estado:
    topFaturadpEstado = vendasGeral.loc[vendasGeral["UF Loja"]==estados,["UF Loja","Descrição do Produto","Total Vendido"]].groupby("Descrição do Produto").sum().sort_values(by="Total Vendido",ascending=False).head(5)
    topFaturadpEstado.plot.barh()
    plt.title(f"TOP 5 PRODUTOS VENDIDOS - {estados}")
    plt.xlabel("Faturamento")
    plt.ylabel("PRODUTOS")
    sns.set()
```

### 2.11 SOMA DA QUANTIDADE VENDIDA POR ESTADO

```python
#SOMA E GRÁFICO DE BARRA HORIZONTAL DA QUANTIDADE VENDIDA TOTAL POR ESTADO
categoriaQuantEstado["Total Vendido"].plot.barh()
plt.title("SOMA DA QUANTIDADE VENDIDA POR ESTADO")
plt.xlabel("QUANTIDADE VENDIDA")
plt.ylabel("ESTADOS")
sns.set()
maiorQuantEstado=(categoriaQuantEstado["Total Vendido"]).to_frame()
display(maiorQuantEstado)
```

```python
#SOMA E GRÁFICO DE BARRA HORIZONTAL DA QUANTIDADE FATURADA TOTAL POR ESTADO
categoriaFaturaEstado["Total Faturado"].plot.barh()
plt.title("SOMA DO FATURAMENTO POR ESTADO")
plt.xlabel("QUANTIDADE VENDIDA")
plt.ylabel("ESTADOS")
sns.set()
maiorFaturaEstado=(categoriaFaturaEstado["Total Faturado"]).to_frame()
display(maiorFaturaEstado)
```
